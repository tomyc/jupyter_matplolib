# A torus

https://scipython.com/book/chapter-7-matplotlib/examples/a-torus/

The parametric description of a torus with radius ***c*** and tube radius ***a*** is
\begin{align}
xyz=(c+acosθ)cosϕ=(c+acosθ)sinϕ=asinθ \\
x=(c+acos⁡θ)cos⁡ϕy=(c+acos⁡θ)sin⁡ϕz=asin⁡θ
\end{align}
for **θ** and **ϕ** each between **0** and **2π**. The code below outputs two views of a torus rendered as a surface plot.

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 100

theta = np.linspace(0, 2.*np.pi, n)
phi = np.linspace(0, 2.*np.pi, n)
theta, phi = np.meshgrid(theta, phi)
c, a = 2, 1
x = (c + a*np.cos(theta)) * np.cos(phi)
y = (c + a*np.cos(theta)) * np.sin(phi)
z = a * np.sin(theta)
```

```python
fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-3,3)
ax1.plot_surface(x, y, z, rstride=5, cstride=5, color='r', edgecolors='w')
ax1.view_init(36, 26)
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_zlim(-3,3)
ax2.plot_surface(x, y, z, rstride=5, cstride=5, color='r', edgecolors='w')
ax2.view_init(0, 0)
ax2.set_xticks([])
plt.show()
```
